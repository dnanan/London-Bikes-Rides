```python
import pandas as pd
```

```python
data = pd.read_csv('london_merged.csv')
```

```python
# Count the uniques values in the weather_code column
data['weather_code'].value_counts()
```

```python
# Count the unique values in the season column
data['season'].value_counts()
```

```python
# Specifying the column names that i want to use
news_cols = {
    'timestamp': 'time',
    'cnt': 'count',
    't1': 'temp_real_C',
    't2': 'temp_feels_like_C',
    'hum': 'humidity_percent',
    'wind_speed': 'wind_speed_kph',
    'weather_code': 'weather',
    'is_holiday': 'holiday',
    'is_weekend': 'weekend',
    'season': 'season'
}

# Renaming the column
data.rename(news_cols, axis = 1, inplace = True)
```

```python
# Changing the humidity values to percentage (ie values between 0 & 1)
data['humidity_percent'] = data['humidity_percent']/100
```

```python
data.head()
```

```python
data.dtypes
```

```python
# Creating a season dictionary so that we can map the integers 0-3 to the actual written values
season_dict = {
    '0': 'spring',
    '1': 'summer',
    '2': 'fall',
    '3': 'winter'
}

# Creating a weather dictionary so that we can map the integers 0-3 to the actual written values
weather_dict = {
    '1': 'Clear',
    '2': 'Scattered Clouds',
    '3': 'Broken Clouds',
    '4': 'Cloudy',
    '7': 'Rain',
    '10': 'Rain With Thunderstorm',
    '26': 'Snowfall'

}
```

```python
# Changing the season column data type to string
data['season'] = data['season'].astype('str')
# Mapping the values 0-3 to the actual written seasons
data['season'] = data['season'].map(season_dict)

# Changing the weather column data type to string
data['weather'] = data['weather'].astype('str')
# Mapping the values to the actual written weathers
data['weather'] = data['weather'].map(weather_dict)
```

```python
# Writing the final data frame to an Excel file that we will use in tableau visualizations. The file will be the 'London_bikes_final"
data.to_excel('London_bikes_final.xlsx', sheet_name= 'Data')
```
